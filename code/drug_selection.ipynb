```python
# combining stable, partial and clinical progressive reponse into incomplete response


# plot histogram for top 4 drugs for breast cancer 


import matplotlib.pyplot as plt
import numpy as np


labels = ['Cyclophosphamide','Doxorubicin', 'Docetaxel','Paclitaxel']
Complete_response = [85, 44, 38, 32]
Incomplete_response = [6, 5, 6, 9]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, Complete_response, width, label='Complete_reponse')
rects2 = ax.bar(x + width/2, Incomplete_response, width, label='Incomplete_response')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('samples')
ax.set_xlabel('drug_name')
ax.set_title('class_imbalance_brca_dfs')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# ax.bar_label(rects1, padding=3)
# ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()

```
