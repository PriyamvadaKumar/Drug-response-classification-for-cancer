```python


from pandas import read_excel
file_name = '/Users/priyamvadakumar/Desktop/research/supplementary_data/BRCA_drug.xlsx' # change it to the name of your excel file
df_brca_drug = read_excel(file_name)
#print(df_brca_drug.head())



print(df_brca_drug)
```

# merge brca_dfs with brca_drug 

```python
brca_dfs_drug=pd.merge(df_brca_dfs,df_brca_drug,on='Unnamed: 0')
#print(brca_dfs_drug.head())

print(brca_dfs_drug)
```

```python
column_names=list(brca_dfs_drug.columns)
print(column_names)
len(column_names)
```

```python
# write csv out



brca_dfs_drug_file=brca_dfs_drug.to_csv(r'/Users/priyamvadakumar/Desktop/research/brca_dfs_drug_file.csv', index = False)
```
